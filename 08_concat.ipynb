### merge

```python
info = {
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}
money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
info_df = pd.DataFrame(info)
money_df = pd.DataFrame(money)
```

```python
# 포링키 중복 요소가 있는 데이터 (여기서는 이름) (컬럼명이 동일)
```

```python
pd.merge(info_df, money_df,how='outer')
```

## merge(how= 'left',  :좌 'right', :우 'outer', :합 'inner'  :교)
## left_on='합칠 기준 칼럼', right_on='합칠 기준 칼럼'

```python
info = {
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}
money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
info_df = pd.DataFrame(info)
money_df = pd.DataFrame(money)
```

```python
pd.merge(info_df, money_df, left_on='이름',right_on='고객명')
```
