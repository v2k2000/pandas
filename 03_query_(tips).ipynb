```python
df['sex'] = df['sex'].astype('object')
conm = df['sex'] == 'Male'
df.loc[conm, 'sex'] = 1 
conf = df['sex'] == 'Female'
df.loc[conf, 'sex'] = 0
df['sex'] = df['sex'].astype('int64')
```

```python
df['smoker'] = df['smoker'].astype('object')
cony = df['smoker'] == 'Yes'
df.loc[cony, 'smoker'] = 1 
conn = df['smoker'] == 'No'
df.loc[conn, 'smoker'] = 0
df['smoker'] = df['smoker'].astype('int64')
```

```python
df['day'] = df['day'].astype('object')
cone = (df['day'] == 'Sun') | (df['day'] == 'Sat')
df.loc[cone, 'day'] = 1 
conw = (df['day'] == 'Thur') | (df['day'] == 'Fri')
df.loc[conw, 'day'] = 0
df['day'] = df['day'].astype('int64')
```

```python
df['time'] = df['time'].astype('object')
cond = (df['time'] == 'Dinner')
df.loc[cond, 'time'] = 1 
conl = (df['time'] == 'Lunch') 
df.loc[conl, 'time'] = 0
df['time'] = df['time'].astype('int64')
```
